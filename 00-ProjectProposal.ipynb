## Data

1. Ideal dataset
To answer our question, our ideal data set would include variables such as the game's name, the corresponding game's tags to describe what kind of game it is, and it would include the game's ratings over time with recorded ratings from 2020 to 2025. Ideally we'd have thousands of games spanning across dozens of different genres and with many different tags, perhaps some even overlapping. This data would ideally be collected by scraping a website, with permission if needed, that has all of the necessary information with as minimal empty values as possible. Ideally, the data would already be tidied and clean. Additionally, it would be stored in multiple tables each connected by the game's title, or a game ID. As of right now we are unsure if it's feasible to keep all of the data in one table, we have not finalized whether to keep tags in separated or contained within a list associated with the game title/ID.

3. Potential real dataset
After searching online, we found that a potential existing dataset for us to use would be on steamdb (https://steamdb.info/). The website already provides documentation for its API, with well documented sections on how to make requests. We do not have to ask for permission for their standard user keys as they are available to everyone and only require a Steam account. For the data that we will be gathering, this level of access is sufficient. The important variables in the dataset that steamdb would grant us include the game's title, a game ID, all of the tags associated with the game, the game's rating over time by average per month, and the game's overall rating as of a specified date.
