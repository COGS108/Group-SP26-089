## Background and Prior Work

The topic we will be focusing on is the general public response to video games developed towards various subjects. More specifically, how a video game’s categorization correlates to the audience’s response to said game as shown through user generated reviews on the Steam platform. We focus on the Steam platform as it is the most widely used PC gaming marketplace, and has games of all genres, styles, and development sizes which would be perfect for the accumulation of data we are researching <a name="valve1"></a>[<sup>5</sup>](#valve1). While games in media are typically taxidermied through single genre classifications, games on Steam are represented by a series of descriptive tags ranging from themes such as “atmospheric” or “horror”, to gameplay like “action” or “adventure”, and even production and style types such as “indie” or “casual”. This process of figuratively stacking classifications provides us the novel opportunity to consider more specific components present in games, and compare data across games with overlapping similarities from various genre groups <a name="tags"></a>[<sup>4</sup>](#tags). A standard user-generated Steam game review is composed of a rating of either positive or negative, and an optional textual review of the game written by the user <a name="valve2"></a>[<sup>6</sup>](#valve2).  In the following paragraphs, I will discuss the concepts of tags and genre in further detail as well as examine familiar research in this field of data study.

An important paper focused on understanding the significance of tags on the Steam platform is A Preliminary Network Analysis on Steam Game Tags: Another Way of Understanding Game Genres by Xiaozhou Li and Boyang Zhang, which discusses the current system of game classification, and the opportunities presented by the uniquely designed Steam tag system. In their research, they consider the user-generated tags on Steam as nodes of a network, with the capability for edges to form between two tags as long as both tags are annotated by the same gamer. Through their study they were able to determine the five most notable game tags, being “indie”, “action”, “adventure”, “singleplayer”, and “casual”, which shows that the tags attached to games span a range of production type, gameplay type, and number of players. This emphasizes the defining point that the standard perception a game from its genre doesn’t properly represent the true interconnectedness games have from one another in the same way that Steam tags do. To further emphasize this idea, the researchers also were able to group tags which fell under the concepts of certain genres, but concluded that it was impossible to find any game on Steam that can be solely classified under a single genre without having some type of element present under another genre<a name="tags"></a>[<sup>4</sup>](#tags). This research helps us understand the unique type of data that game tags presents to us, while also proving the difference our research would have over the genre-specific research to be defined in the following section.

Facebook IQ, in collaboration with external research organizations, conducted a global study of 13,412 mobile gamers across 11 countries segmented across 27 different mobile gaming genres and subgenres to explore their gaming behaviors while also utilizing facebook’s internal app events data to uncover behavioral insights across mobile gaming genres. Through their research, they discovered that the specific genre preferences vary significantly across countries, but some notable genres include matching puzzle games, card/casino games, racing games, and action RPG/MMORPG games which gives us a good insight into genre preferences in a gaming marketplace outside of Steam. Furthermore, the study also indicates that very few gamers play only one genre, with players in the US, UK, and South Korea reportedly playing an average of two distinct genres in a given month, while Japanese play three. This is important to note for our own study as it shows that the population of user responses to games on Steam aren’t evenly distributed between tags, with the expectation that single users may have reviewed multiple games of different categories within our data. Finally, we can learn that the study we will be conducting would be important, as across nine surveyed markets, nearly half of player responses say genre is important in their decision, greatly exceeding other considerations <a name="facebook"></a>[<sup>1</sup>](#facebook). 

The closest research we could find to our own focus of analyzing user response to games of different tags comes from the paper Prediction of the Most Popular Game Tags on Steam under the Influence of Covid-19 Based on Machine Learning and Natural Language Processing, in which the authors applied various machine learning algorithms to Steam’s game tags to create a prediction model. Unfortunately, due to the research’s minimal scope of data only relating to the Covid-19 pandemic, they were unable to find conclusive evidence of tag correlation using their various methods <a name="prediction"></a>[<sup>3</sup>](#prediction). From this, we can learn that for our own research we would need to accumulate a significant amount of data to achieve any type of valid conclusion. That being said, the paper From Fadas to Classics which analyzes the evolution of Steam Tags throughout the years, concludes that tag popularity tends to fluctuate with time, with many trends lasting about four years <a name="fads"></a>[<sup>2</sup>](#fads) which is important for us to keep in mind when accumulating our own data.
