```python
def CreateDataForGame(row, rank_num):

    #See if the game is in the dictionary
    if int(row['AppID']) not in review_data:
        return np.NaN
        
    #Import df from the game review dictionary
    counts = review_data[int(row['AppID'])].groupby('timestamp_created').value_counts()
    #Get the year index
    years = list(set(counts.index.get_level_values(0)))
    years = list(map(str, years))
    twenty_four_ratings = np.NaN
    twenty_five_ratings = np.NaN
    twenty_six_ratings = np.NaN
    if '2024' in years:
        starting_index = years.index('2024')
        twenty_four_ratings = counts.iloc[(starting_index * 2):((starting_index * 2) + 2)]
    if '2025' in years:
        starting_index = years.index('2025')
        twenty_five_ratings = counts.iloc[(starting_index * 2):((starting_index * 2) + 2)]
    if '2026' in years:
        starting_index = years.index('2026')
        twenty_six_ratings = counts.iloc[(starting_index * 2):((starting_index * 2) + 2)]
    return pd.Series({"appID": row['AppID'], "2024_Ratings": twenty_four_ratings, "2025_Ratings": twenty_five_ratings, "2026_Ratings": twenty_six_ratings, "rank": rankNum})
```
