## Abstract

This project sought to examine if a relationship between the apperance of specific Steam game tags and community reception, and ratings, existed between 2010 and early 2026. Understanding this trend is important because it is a representation of how impactful having specific tags attached to games is toward their rating and reception. While genres give you a general idea of what to expect from the game, tags give specifics about the feel and vibe of the game which may make players have stronger feelings toward certain characteristic tags as opposed to others.

To evaluate this trend, we found and cleaned a recently scraped dataset (as of March 2026) of Steam games and their information from Kaggle, and scraped Steam for the reviews to compare the percentage of positive ratings per year by associating the tags with the game they're attached to. First, we visualized the abundance of unique tags with a WordCloud followed by bar graphs to display the distributions of tag appearances. Next we used box plots to examine the distribution of review scores per tag and to identify any outliers. Bar graphs were used to display raw average review rating per tag using game's average percent score, and to identify any connection between tags, we used a bar grpah to display the appearances of pairs of tags. Finally, line plots were used to display positive ratings over time per tag and groups of similar/connected tags.

Our results showed that all of our tags showed similar increases and decreases in reception and rating during the same time frames. Additionally, there are no specific tags, of the ones we showed, that corellate with more positive or negative reception because of the similarity in increases and decreases. Overall, our analysis demonstrated that there does exist a close relationship between the sentiment of Steam games including one of our top 10 tags and community reception rating on Steam between 2010 and early 2026, and reception changed significantly over time with similar trends appearing across every tag.

## Background and Prior Work

The topic we will be focusing on is the general public response to video games developed towards various subjects. More specifically, how a video game’s categorization correlates to the audience’s response to said game as shown through user generated reviews on the Steam platform. We focus on the Steam platform as it is the most widely used PC gaming marketplace, and has games of all genres, styles, and development sizes which would be perfect for the accumulation of data we are researching <a name="valve1"></a>[<sup>5</sup>](#valve1). While games in media are typically taxidermied through single genre classifications, games on Steam are represented by a series of descriptive tags ranging from themes such as “atmospheric” or “horror”, to gameplay like “action” or “adventure”, and even production and style types such as “indie” or “casual”. This process of figuratively stacking classifications provides us the novel opportunity to consider more specific components present in games, and compare data across games with overlapping similarities from various genre groups <a name="tags"></a>[<sup>4</sup>](#tags). A standard user-generated Steam game review is composed of a rating of either positive or negative, and an optional textual review of the game written by the user <a name="valve2"></a>[<sup>6</sup>](#valve2).  In the following paragraphs, I will discuss the concepts of tags and genre in further detail as well as examine familiar research in this field of data study.

An important paper focused on understanding the significance of tags on the Steam platform is A Preliminary Network Analysis on Steam Game Tags: Another Way of Understanding Game Genres by Xiaozhou Li and Boyang Zhang, which discusses the current system of game classification, and the opportunities presented by the uniquely designed Steam tag system. In their research, they consider the user-generated tags on Steam as nodes of a network, with the capability for edges to form between two tags as long as both tags are annotated by the same gamer. Through their study they were able to determine the five most notable game tags, being “indie”, “action”, “adventure”, “singleplayer”, and “casual”, which shows that the tags attached to games span a range of production type, gameplay type, and number of players. This emphasizes the defining point that the standard perception a game from its genre doesn’t properly represent the true interconnectedness games have from one another in the same way that Steam tags do. To further emphasize this idea, the researchers also were able to group tags which fell under the concepts of certain genres, but concluded that it was impossible to find any game on Steam that can be solely classified under a single genre without having some type of element present under another genre<a name="tags"></a>[<sup>4</sup>](#tags). This research helps us understand the unique type of data that game tags presents to us, while also proving the difference our research would have over the genre-specific research to be defined in the following section.

Facebook IQ, in collaboration with external research organizations, conducted a global study of 13,412 mobile gamers across 11 countries segmented across 27 different mobile gaming genres and subgenres to explore their gaming behaviors while also utilizing facebook’s internal app events data to uncover behavioral insights across mobile gaming genres. Through their research, they discovered that the specific genre preferences vary significantly across countries, but some notable genres include matching puzzle games, card/casino games, racing games, and action RPG/MMORPG games which gives us a good insight into genre preferences in a gaming marketplace outside of Steam. Furthermore, the study also indicates that very few gamers play only one genre, with players in the US, UK, and South Korea reportedly playing an average of two distinct genres in a given month, while Japanese play three. This is important to note for our own study as it shows that the population of user responses to games on Steam aren’t evenly distributed between tags, with the expectation that single users may have reviewed multiple games of different categories within our data. Finally, we can learn that the study we will be conducting would be important, as across nine surveyed markets, nearly half of player responses say genre is important in their decision, greatly exceeding other considerations <a name="facebook"></a>[<sup>1</sup>](#facebook). 

The closest research we could find to our own focus of analyzing user response to games of different tags comes from the paper Prediction of the Most Popular Game Tags on Steam under the Influence of Covid-19 Based on Machine Learning and Natural Language Processing, in which the authors applied various machine learning algorithms to Steam’s game tags to create a prediction model. Unfortunately, due to the research’s minimal scope of data only relating to the Covid-19 pandemic, they were unable to find conclusive evidence of tag correlation using their various methods <a name="prediction"></a>[<sup>3</sup>](#prediction). From this, we can learn that for our own research we would need to accumulate a significant amount of data to achieve any type of valid conclusion. That being said, the paper From Fadas to Classics which analyzes the evolution of Steam Tags throughout the years, concludes that tag popularity tends to fluctuate with time, with many trends lasting about four years <a name="fads"></a>[<sup>2</sup>](#fads) which is important for us to keep in mind when accumulating our own data.

### Steam Game Tags EDA

Wow that is a lot of tags! Now we will look at the top 10 most prevalent tags, followed by the tags specified in our original research question.

```python
# pure top 10
test = dict(Counter(flattened_tags))
dfr = (pd.DataFrame.from_dict(test, orient='index')).set_axis(['count'], axis=1)
dfr.reset_index(level=0, inplace=True)
dfr.rename(columns={"index": "tag", 'count': 'count'})
top_10_by_tag_ct = (dfr.sort_values(by = ['count'], ascending = False)).head(10)

f1 = sns.barplot(top_10_by_tag_ct, x = 'count', y = 'index', hue = 'index')
f1.set_title("Top 10 Steam Tags By Appearance")
f1.set_xlabel('Tag')
f1.set_ylabel('Count')
f1.tick_params(axis='x', labelsize=10)
plt.show()

# our top 10
our_tags_df  = dfr[dfr['index'].isin(our_tags)]
our_tags_df  = (our_tags_df.sort_values(by = ['count'], ascending = False))

f2 = sns.barplot(our_tags_df, x = 'count', y = 'index', hue = 'index')
f2.set_title("Our Top 10 Steam Tags By Appearance")
f2.set_xlabel('Tag')
f2.set_ylabel('Count')
f2.tick_params(axis='x', labelsize=10)
plt.show()
```

Of the 10 tags we initially chose, 7 were present in the top 10 by appearance! This was interesting to us because we had chosen our top 10 tags based on what Steam had proposed were its 10 most popular tags. 

Next we will look at raw average percent score for every one of our original 10 tags. We'll start with adding a new column that only has the first 3 tags from each game, we'll discuss later why we chose 3.

We'll consider two cases:

Case 1: all 10 tags from original data set are considered

Case 2: only the first 3 tags from the original data set are considered

We will save two dataframes for each case

```python
# adds a new column that holds the first 3 tags from the All_Tags column
tags_list = data['All_Tags'].values.tolist()

i = 0
for x in tags_list:
    tags_list[i] = x.split(';')
    tags_list[i] = tags_list[i][:3]
    i += 1

# cut down data to only ones which is within the top ten tags
data['Top_3_Tags'] = tags_list
top_data = data[data.apply(lambda x: (any(substring in x['All_Tags'] for substring in our_tags)), axis=1)]
top_three_data = data[data.apply(lambda x: (any(substring in x['Top_3_Tags'] for substring in our_tags)), axis=1)]
```

We'll visualize the distribution of review scores per tag.

```python
# all 10 tags boxplot
ind_tag_data = pd.DataFrame(columns=['tag_name', 'review_score'])
for tag_name in our_tags:
    for index, row in data.iterrows():
        if (tag_name in row['All_Tags']):
            ind_tag_data.loc[len(ind_tag_data)] = [tag_name, row['Review_Score_Pct']]

sns.boxplot(y="tag_name", x="review_score", data=ind_tag_data)
plt.title('Distribution of Review Scores per Tag (all 10 tags considered)')
plt.xlabel('Review Score')
plt.ylabel('Tag')
plt.show()

# first 3 tags boxplot
top_data_tags = pd.DataFrame(columns=['tag_name', 'review_score'])
for tag_name in our_tags:
    for index, row in data.iterrows():
        if (tag_name in row['Top_3_Tags']):
            top_data_tags.loc[len(top_data_tags)] = [tag_name, row['Review_Score_Pct']]

sns.boxplot(y="tag_name", x="review_score", data=top_data_tags)
plt.title('Distribution of Review Scores per Tag (only first 3 tags considered)')
plt.xlabel('Review Score')
plt.ylabel('Tag')
plt.show()
```

The distribution of review scores had significant outliers when all 10 tags were considered, and it was interesting to us that a major portion of the outliers disappeared when we only took the first 3 tags of each game into account when compared to our tags list. We believe this to be because the first 3 tags hold significantly more weight (i.e connection to the game and the number of users that assigned the tag to the game), therefore the games and its reviews and scores are more closely connected to the first 3 tags as opposed to all 10 equally.

Finally we'll look at the raw average review score per tag using the game's cumulative review score percent. Similar to the boxplot, we will consider the two different cases!

```python
# all 10 tags consicered
avg_per_per_tag = pd.DataFrame(columns=["tag_name", "avg"]) 

for tag_name in our_tags:
    sum_of_avgs = 0
    num_of_games = 0
    for index, row in data.iterrows():
        if (tag_name in row['All_Tags']):
            sum_of_avgs += row['Review_Score_Pct']
            num_of_games += 1
    avg_per_per_tag.loc[len(avg_per_per_tag)] = [tag_name, sum_of_avgs / num_of_games]

sns.barplot(data=avg_per_per_tag, y="tag_name", x='avg')
plt.title('Raw Average Review Score per Tag (all 10 tags considered)')
plt.xlabel('Review Score')
plt.ylabel('Tag')
plt.show()

# only first 3 tags considered
three_avg = pd.DataFrame(columns=["tag_name", "avg"]) 

for tag_name in our_tags:
    sum_of_avgs = 0
    num_of_games = 0
    for index, row in data.iterrows():
        if (tag_name in row['Top_3_Tags']):
            sum_of_avgs += row['Review_Score_Pct']
            num_of_games += 1
    three_avg.loc[len(three_avg)] = [tag_name, sum_of_avgs / num_of_games]

sns.barplot(data=three_avg, y="tag_name", x='avg')
plt.title('Raw Average Review Score per Tag (only first 3 tags considered)')
plt.xlabel('Review Score')
plt.ylabel('Tag')
plt.show()
```

Interestingly, the pure raw average review scores per tag when considered if the tag was present at all versus only in the first 3 tags showed no significant changes. Additionally, all of our tags showed raw average review score within the 78-85% range with the only clear "leader" being the '2D' tag.

With 437 unique tags, we wanted to see what pairs of tags most commonly appeared together in games' tags lists. We used Counter to make looking for combinations easier and more streamlined.

```python
pair_counts = Counter()

data['Tags_List'] = data['All_Tags'].str.split(';')

for tags in data['Tags_List']:
    pair_counts.update(combinations(sorted(tags), 2))

pair_dict = dict(pair_counts)
top_pairs = {}

for key, value in pair_dict.items():
    if any(item in our_tags for item in key):
        top_pairs.update({key:value})
top_pairs = Counter(top_pairs)
top = top_pairs.most_common(20)

plot_df = pd.DataFrame(top, columns=["pair", "count"])
plot_df["pair"] = plot_df["pair"].apply(lambda x: " + ".join(x))

sns.barplot(data=plot_df, y="pair", x="count")
plt.title("Most Common Steam Tag Pairs Where at Least 1 Tag is From our List of Tags")
plt.xlabel('Count')
plt.ylabel('Tag Pairs')
plt.show()
```

Of the top 20 most common tag pairs, almost half of them (9 to be exact) include 2 tags from our chosen top 10 tags list! This shows that a lot of our tags are connected in one way or another, and that trying to plot review scores and trends for each individual tag might skew the data such that one tags scores would overpower anothers. We take this into account in the lattere falf of the next analysis subsection.

In conclusion, we discovered that through the overlap of games having several of our chosen ten tags, the cumulative review average per tag did not have a significant difference between each other, but the total average score did cumulatively fluctuate for most tags per year. The combination of several tags to describe a Steam game was an expected outcome, and follows through with Li and Zhang’s study on how tags span a range of production type, gameplay type, and number of players to represent a game <a name="tags"></a>[<sup>4</sup>](#tags). This was directly observed when testing for tag pairs, with the top two pairs being “Action + Singleplayer” and “Adventure + Singleplayer”, both representing a gameplay type with a number of players. The close averages between tags also relates with our proposal’s expectation derived from the Facebook IQ paper, with users potentially playing and rating games from several tags on Steam, causing a dispersal of the population’s direct affect on an individual tag. It is interesting to note though that compared to Facebook’s highest frequency findings of matching puzzle games, card/casino games, racing games, and action RPG/MMORPG games, our test of the top ten Steam tags rather finds only one direct connection, being action games. This could just be due to the fact that as mentioned previously, Steam tags reflect differently than a standard genre system which changes the standard classification method of each game <a name="facebook"></a>[<sup>1</sup>](#facebook). One of the most surprising results from our study was the closeness of user scores regardless of tags per year, as while there is a mild fluctuation present, it is too small of a difference to build on the conclusion that specific tag popularity fluctuates with time such as what was said in the paper From Fadas to Classics <a name="fads"></a>[<sup>2</sup>](#fads). 
There are several limitations with our results to be considered. First and foremost being the fact that we were only able to scrape the top ~5400 of Steam’s over hundred thousand games with enough reviews to be considered viable for our analysis. We also had to cap the number of reviews per game at ten thousand, to get our data in a realistic amount of time without disrupting the API. Through the API we were only able to separate the given data into years, which hindered us from potentially considering review fluctuations on a monthly basis or beyond. Finally, due to the need for games to have at least 200 reviews per year to be suitable enough to graph, we could not properly do random sampling and had to rely on the top rated games instead. Being capable of using the yearly average of all the reviews per game rather than a maximum of ten thousand, and being able to correlate reviews to months rather than years would improve confidence in our results. Furthermore, if we were able to reliably scrape and analyze each written review for each game rather than using a boolean positive or negative score, we could’ve utilized sentiment analysis to dive deeper into the true level of response to each tag. There are a few considered paths to take if we were to investigate this further, such as creating an interactive program to select and visualize chosen tag averages to aid in finding unique differences, and looking into the addition of tags on Steam and how those additions shifted tag presence in games on Steam throughout the years.
